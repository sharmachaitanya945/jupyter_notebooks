## Exercise 1: Create a Pandas Series

**Task:** Create a Pandas Series from a list of numbers and print its values.

```python
# Create a Pandas Series
numbers = [10, 20, 30, 40, 50]
series = pd.Series(numbers)
print(series)
```

## Exercise 2: Exploring DataFrame Basics

**Task:** Create a DataFrame from a dictionary and display its contents.

```python
# Creating a DataFrame from a dictionary
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35],
        'City': ['New York', 'San Francisco', 'Los Angeles']}
df = pd.DataFrame(data)
print(df)
```

## Exercise 3: DataFrame Information

**Task:** Display basic information about the DataFrame using `.info()` and `.describe()`.

```python
# Display DataFrame information
df.info()
df.describe()
```

## Exercise 4: Selecting Specific Columns

**Task:** Select the 'Name' column from the DataFrame.

```python
# Selecting a specific column
name_column = df['Name']
print(name_column)
```

## Exercise 5: Filtering Data

**Task:** Filter the DataFrame to show only rows where Age is greater than 28.

```python
# Filtering the DataFrame
filtered_df = df[df['Age'] > 28]
print(filtered_df)
```
